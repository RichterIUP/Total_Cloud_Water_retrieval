```python
cloud_file = "cbh_{:04d}{:02d}{:02d}_{:02d}{:02d}{:02d}_{:.2f}_{:.2f}.nc".format(date.year, \
                                                                                                  date.month, \
                                                                                                  date.day, \
                                                                                                  date.hour, \
                                                                                                  date.minute, \
                                                                                                  date.second, \
                                                                                                  int(lat+1), \
                                                                                                  int(lat-1), \
                                                                                                  int(lon+1), \
                                                                                                  int(lon-1))
if not os.path.exists(cloud_file):
    retrieval.download_cloud_position(cloud_file)
retrieval.read_cloud_height(cloud_file)
```

```python
retrieval.prepare_retrieval()
mcp = retrieval.get_mcp()
print(mcp)
```

```python
retrieval.set_mcp(np.array([0.0, 0.0, 5.0, 40.0]))
retrieval.run_LBLDIS(clear_sky=True)
retrieval.set_mcp(np.array([0.1, 0.1, 5.0, 20.]))
```

```python
for i in range(10):
    retrieval.run_LBLDIS(clear_sky=False)
    retrieval.run_LBLDIS_deriv()
    retrieval.calculate_cloud_emissivity()
    retrieval.plot_emissivity()
    #retrieval.plot_spectrum()
    retrieval.jacobian()
    retrieval.calc_chi_2_and_residuum()
    if retrieval.convergence():
        break
    retrieval.adjust_lm()
    retrieval.iteration()
    retrieval.adjust_mcp()
    print(retrieval.get_mcp())
```
